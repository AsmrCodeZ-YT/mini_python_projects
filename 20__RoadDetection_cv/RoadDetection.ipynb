```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv.imread("pp.jpg", 0)
```

```python
# remove noise from picture
image_gaussin = cv.GaussianBlur(image ,(3,3) ,0)

laplacian = cv.Laplacian(image,cv.CV_64F)

#orginalPic
plt.subplot(2,2,1) ,plt.imshow(image ,cmap="gray")
plt.title("originalPic") ,plt.xticks([]), plt.yticks([])
#laplacianPic
plt.subplot(2,2,2) ,plt.imshow(laplacian ,cmap="gray")
plt.title("laplacianPic") ,plt.xticks([]), plt.yticks([])
```

```python
# creat a custom kernel without Deeplearning , python trick

sobel_x = np.array([[-1,0,1],
                    [-2,0,2],
                    [-1,0,1]])

# sobel_y 
sobel_y = np.array([[-1,2,1],
                    [0,0,0],
                    [-1,-2,1]])
```

```python
filtered_image_y = cv.filter2D(image ,-1 ,sobel_y)
filtered_image_x = cv.filter2D(image ,-1 ,sobel_x)



#plot
plt.subplot(2,2,1) ,plt.imshow(image )
plt.title("originalPic") ,plt.xticks([]), plt.yticks([])
plt.subplot(2,2,2) ,plt.imshow(laplacian ,cmap="gray")
plt.title("laplacian") ,plt.xticks([]), plt.yticks([])
plt.subplot(2,2,3) ,plt.imshow(filtered_image_x ,cmap="gray")
plt.title("filtered_image_x") ,plt.xticks([]), plt.yticks([])
plt.subplot(2,2,4) ,plt.imshow(filtered_image_y ,cmap="gray")
plt.title("filtered_image_y") ,plt.xticks([]), plt.yticks([])
plt.show()
```
